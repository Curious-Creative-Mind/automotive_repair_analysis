### 2. Data Cleaning:

```python
df = df.drop_duplicates(subset='VIN')
```

```python
df = df.drop(columns=['CAMPAIGN_NBR','COMPLAINT_CD_CSI','TRANSACTION_CNTR'])
```

```python
# converting free text columns into lower case and trimming text for further operations

df['CORRECTION_VERBATIM'] = df['CORRECTION_VERBATIM'].astype(str).str.lower().str.strip()
df['CUSTOMER_VERBATIM'] = df['CUSTOMER_VERBATIM'].astype(str).str.lower().str.strip()
```

```python
df = df.reset_index(drop=True)
```

```python
# converting string values into float for cost calculation columns

cols_to_convert = ['REPORTING_COST', 'TOTALCOST', 'LBRCOST']

for col in cols_to_convert:
    df[col] = df[col].str.replace(',', '', regex=False).astype(float)
```

```python
# Imputing value in blank records of column TOTALCOST with corresponding values from REPORTING_COST column as mostly they have same values.

df['TOTALCOST'] = df['TOTALCOST'].fillna(df['REPORTING_COST'])
```

```python
# Filling null values with appropriate data

df['CAUSAL_PART_NM'] = df['CAUSAL_PART_NM'].fillna("N/A")
df['PLANT'] = df['PLANT'].fillna("SIL")
```

### Generating tags/features from free text available

```python
# Defining Keywords for each category for col CUSTOMER_VERBATIM

failure = ['works','working']
poor_qlty_comp = ['return', 'peeling', 'delaminating']
loose_fit = ['coming apart', 'coming off', 'steering wheel']

# Function for tagging
def tag_issue(text):
    
    if any(word in text for word in failure):
        return 'Failure'
    elif any(word in text for word in poor_qlty_comp):
        return 'Poor Quality Component'
    elif any(word in text for word in loose_fit):
        return 'Component Loose Fitting'
    else:
        return 'Other'
        
# Apply the function to create a new column
df['ISSUE_CATEGORY'] = df['CUSTOMER_VERBATIM'].apply(tag_issue)
```

```python
# Defining Keywords for each category for col CORRECTION_VERBATIM

repaired = ['adjusted', 'reconnected', 'fix', 'installed', 'reassemble', 'return'] 
replaced = ['replaced', 'replace', 'new wheel']

# Function for tagging
def tag_sol_actn(text):
    
    if any(word in text for word in repaired):
        return 'Repaired'
    elif any(word in text for word in replaced):
        return 'Component Replaced'
    else:
        return 'Other'
        
# Apply the function to create a new column
df['SOLUN_ACTN_CATEGORY'] = df['CORRECTION_VERBATIM'].apply(tag_sol_actn)
```
